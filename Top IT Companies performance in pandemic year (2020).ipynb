# Finance Data EDA

```python
from pandas_datareader import data, wb
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import datetime
%matplotlib inline
```

```python
start = datetime.datetime(2020, 1, 1)
end = datetime.datetime(2021, 1, 1)
```

```python
#Getting the data from yahoo finance
# HCL
HCL = data.DataReader("HCLTECH.NS", 'yahoo', start, end)

# Infosys
INFOSYS = data.DataReader("INFY.NS", 'yahoo', start, end)

# Wipro
WIPRO = data.DataReader("WIPRO.NS", 'yahoo', start, end)

# Mphasis
MPHASIS = data.DataReader("MPHASIS.NS", 'yahoo', start, end)

# Tech Mahindra
TM = data.DataReader("TECHM.NS", 'yahoo', start, end)

# LARSEN AND TOUBRO
LNT = data.DataReader("LTI.NS", 'yahoo', start, end)
```

```python
#df=data.DataReader(['HCL','INFOSYS','WIPRO','MPHASIS','TM','LNT'],'yahoo',start,end)
```

```python
company=['HCL','INFOSYS','WIPRO','MPHASIS','TM','LNT']
```

```python
stocks=pd.concat([HCL,INFOSYS,WIPRO,MPHASIS,TM,LNT],axis=1,keys=company)
```

```python
stocks.columns.names=['Company','Stock Info']
```

```python
stocks.head(10)
```

```python
stocks.xs(key='Close',axis=1,level='Stock Info').max
```

```python
returns = pd.DataFrame()
for tick in company:
    returns[tick+'Return']=stocks[tick]['Close'].pct_change()
returns.head()
```
